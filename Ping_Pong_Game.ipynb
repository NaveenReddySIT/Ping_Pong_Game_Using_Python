# Building the main code frame

```python
import pygame
import sys
import random

# Particle class
class Particle:
    def __init__(self, x, y, color):
        self.x = x
        self.y = y
        self.color = color
        self.size = random.randint(2, 5)
        self.speed_x = random.uniform(-1, 1)
        self.speed_y = random.uniform(-1, 1)
        self.lifetime = 30

    def update(self):
        self.x += self.speed_x
        self.y += self.speed_y
        self.lifetime -= 1

    def draw(self, screen):
        pygame.draw.circle(screen, self.color, (int(self.x), int(self.y)), self.size)

particles = []

score_player = 0
score_opponent = 0

def create_particles(position, color, count):
    for _ in range(count):
        x = position[0]
        y = position[1]
        particle = Particle(x, y, color)
        particles.append(particle)

def update_particles():
    particles_to_remove = []

    for particle in particles:
        if particle.lifetime <= 0:
            particles_to_remove.append(particle)
        else:
            particle.update()

    for particle in particles_to_remove:
        particles.remove(particle)

def draw_particles(screen):
    for particle in particles:
        particle.draw(screen)

# Font setup
pygame.font.init()
font = pygame.font.Font(None, 70)

# Game setup
pygame.init()
clock = pygame.time.Clock()

screen_width = 1100
screen_height = 600
screen = pygame.display.set_mode((screen_width, screen_height))
pygame.display.set_caption("Ping Pong Party")

background_color = pygame.Color('black')
light_grey = (200, 200, 200)

# Game mode setup
game_mode = None

# Title display
title_timer = 2 * 60  # 10 seconds at 60 FPS
show_title = True

# Button class
class Button:
    def __init__(self, x, y, width, height, text, action=None):
        self.rect = pygame.Rect(x, y, width, height)
        self.color = light_grey
        self.text = text
        self.action = action

    def draw(self, screen):
        pygame.draw.rect(screen, self.color, self.rect)
        font_size = 40
        font = pygame.font.Font(None, font_size)
        text = font.render(self.text, True, background_color)
        text_rect = text.get_rect(center=self.rect.center)
        screen.blit(text, text_rect)

    def handle_event(self, event):
        if event.type == pygame.MOUSEBUTTONDOWN:
            if self.rect.collidepoint(event.pos):
                if self.action is not None:
                    self.action()

# Button constants
BUTTON_WIDTH = 200
BUTTON_HEIGHT = 70
BUTTON_SPACING = 20

# Single player mode function
def single_player_mode():
    global game_mode
    game_mode = "single"

# Two players mode function
def two_players_mode():
    global game_mode
    game_mode = "two_players"

# Single player button setup
single_player_button_x = (screen_width - BUTTON_WIDTH) / 2
single_player_button_y = (screen_height - BUTTON_HEIGHT) / 2 - BUTTON_HEIGHT - BUTTON_SPACING
single_player_button = Button(single_player_button_x, single_player_button_y, BUTTON_WIDTH, BUTTON_HEIGHT, "Single Player", single_player_mode)

# Two players button setup
two_players_button_x = (screen_width - BUTTON_WIDTH) / 2
two_players_button_y = (screen_height - BUTTON_HEIGHT) / 2 + BUTTON_SPACING
two_players_button = Button(two_players_button_x, two_players_button_y, BUTTON_WIDTH, BUTTON_HEIGHT, "2 Players", two_players_mode)

# Player speed variables
player_speed = 0
opponent_speed = 0

# Player rectangles
player_width = 10
player_height = 140
player_x = screen_width - player_width - 20
player_y = screen_height / 2 - player_height / 2
player = pygame.Rect(player_x, player_y, player_width, player_height)

opponent_width = 10
opponent_height = 140
opponent_x = 10
opponent_y = screen_height / 2 - opponent_height / 2
opponent = pygame.Rect(opponent_x, opponent_y, opponent_width, opponent_height)

# Ball variables
ball_radius = 10
ball_x = screen_width / 2
ball_y = screen_height / 2
ball_speed_x = random.choice([-4, 4])
ball_speed_y = random.choice([-4, 4])
ball = pygame.Rect(ball_x - ball_radius, ball_y - ball_radius, ball_radius * 2, ball_radius * 2)

# Game loop
while True:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            sys.exit()
        if game_mode is None:
            single_player_button.handle_event(event)
            two_players_button.handle_event(event)
        if game_mode == "two_players":
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_UP:
                    player_speed -= 7
                if event.key == pygame.K_DOWN:
                    player_speed += 7
                if event.key == pygame.K_w:
                    opponent_speed -= 7
                if event.key == pygame.K_s:
                    opponent_speed += 7
            if event.type == pygame.KEYUP:
                if event.key == pygame.K_UP:
                    player_speed += 7
                if event.key == pygame.K_DOWN:
                    player_speed -= 7
                if event.key == pygame.K_w:
                    opponent_speed += 7
                if event.key == pygame.K_s:
                    opponent_speed -= 7
        elif game_mode == "single":
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_UP:
                    player_speed -= 7
                if event.key == pygame.K_DOWN:
                    player_speed += 7
            if event.type == pygame.KEYUP:
                if event.key == pygame.K_UP:
                    player_speed += 7
                if event.key == pygame.K_DOWN:
                    player_speed -= 7

    if show_title:
        screen.fill(background_color)
        title_text = font.render("Ping Pong Party", True, light_grey)
        title_rect = title_text.get_rect(center=(screen_width / 2, screen_height / 2))
        screen.blit(title_text, title_rect)
        title_timer -= 1
        if title_timer <= 0:
            show_title = False

    elif game_mode is None:
        screen.fill(background_color)
        mode_text = font.render("SELECT MODE TO PLAY", True, light_grey)
        mode_text_rect = mode_text.get_rect(center=(screen_width / 2, screen_height / 2 - 50))
        screen.blit(mode_text, mode_text_rect)

        single_player_button.draw(screen)
        two_players_button.draw(screen)

    else:
        # Ball animation function
        def ball_animation():
            global score_player, score_opponent, ball_speed_x, ball_speed_y
            ball.x += ball_speed_x
            ball.y += ball_speed_y

            if ball.top <= 0 or ball.bottom >= screen_height:
                ball_speed_y *= -1

            if ball.colliderect(player) and ball_speed_x > 0:
                ball_speed_x *= -1
                create_particles((ball.left, ball.y), light_grey, 10)

            if ball.colliderect(opponent) and ball_speed_x < 0:
                ball_speed_x *= -1
                create_particles((ball.right, ball.y), light_grey, 10)

            if ball.left <= 0:
                score_player += 1
                ball_restart()

            if ball.right >= screen_width:
                score_opponent += 1
                ball_restart()

        def ball_restart():
            global ball_speed_x, ball_speed_y, ball_x, ball_y
            ball.center = (screen_width / 2, screen_height / 2)
            ball_speed_y *= random.choice([-1, 1])
            ball_speed_x *= random.choice([-1, 1])

        def player_animation():
            player.y += player_speed

            if player.top <= 0:
                player.top = 0
            if player.bottom >= screen_height:
                player.bottom = screen_height

        def opponent_ai():
            if opponent.top < ball.y:
                opponent.top += 4
            if opponent.bottom > ball.y:
                opponent.bottom -= 4
            if opponent.top <= 0:
                opponent.top = 0
            if opponent.bottom >= screen_height:
                opponent.bottom = screen_height

        # Game logic
        player_animation()

        if game_mode == "single":
            opponent_ai()

        ball_animation()
        update_particles()

        # Drawing
        screen.fill(background_color)
        pygame.draw.rect(screen, light_grey, player)
        pygame.draw.rect(screen, light_grey, opponent)
        pygame.draw.ellipse(screen, light_grey, ball)
        pygame.draw.aaline(screen, light_grey, (screen_width / 2, 0), (screen_width / 2, screen_height))

        # Draw scores
        score_player_text = font.render(str(score_player), True, light_grey)
        score_opponent_text = font.render(str(score_opponent), True, light_grey)
        score_player_rect = score_player_text.get_rect(midleft=(screen_width / 2 + 40, screen_height / 2))
        score_opponent_rect = score_opponent_text.get_rect(midright=(screen_width / 2 - 40, screen_height / 2))
        screen.blit(score_player_text, score_player_rect)
        screen.blit(score_opponent_text, score_opponent_rect)

        # Draw particles
        draw_particles(screen)

    pygame.display.flip()
    clock.tick(60)
```
